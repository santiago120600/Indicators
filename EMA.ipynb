# Exponential Moving Average (EMA)
The Exponential Moving Average is calculated as:

$$
\text{EMA}_t = \alpha \cdot P_t + (1 - \alpha) \cdot \text{EMA}_{t-1}
$$

where:
- $\text{EMA}_t$ = EMA at time $t$
- $P_t$ = price/value at time $t$
- $\alpha = \frac{2}{N+1}$ = smoothing factor
- $N$ = number of periods
- $\text{EMA}_{t-1}$ = previous period's EMA

### Example
Last 5 closing prices are: 20, 22, 21, 23, 24

#### 1. Compute the smoothing factor

$$
\alpha = \frac{2}{N + 1}
$$

Given:

$$
N = 5
$$

$$
\alpha = \frac{2}{6} = 0.3333
$$

$$
1 - \alpha = 0.6667
$$

---

#### 2. Price series

$$
P = [20, 22, 21, 23, 24]
$$

---

#### 3. Initial EMA

We use the first value as the initial EMA:

$$
\text{EMA}_1 = 20
$$

---

#### 4. Step-by-step calculation

##### Step 1

$$
\text{EMA}_1 = 20
$$

---

##### Step 2

$$
\text{EMA}_2 = 0.3333 \cdot 22 + 0.6667 \cdot 20
$$

$$
= 7.3333 + 13.3334 = 20.6667
$$

---

##### Step 3

$$
\text{EMA}_3 = 0.3333 \cdot 21 + 0.6667 \cdot 20.6667
$$

$$
= 7 + 13.7778 = 20.7778
$$

---

##### Step 4

$$
\text{EMA}_4 = 0.3333 \cdot 23 + 0.6667 \cdot 20.7778
$$

$$
= 7.6667 + 13.8519 = 21.5186
$$

---

##### Step 5

$$
\text{EMA}_5 = 0.3333 \cdot 24 + 0.6667 \cdot 21.5186
$$

$$
= 8 + 14.3457 = 22.3457
$$

---

##### 5. Final Results

| t | Price | EMA     |
|---|------|---------|
| 1 | 20   | 20.0000 |
| 2 | 22   | 20.6667 |
| 3 | 21   | 20.7778 |
| 4 | 23   | 21.5186 |
| 5 | 24   | 22.3457 |


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.random.randint(0, 100, 30).tolist()
df = pd.DataFrame(data, columns=['Close'])
df
```

```python
df.plot(title='Close Price')
plt.show()
```

```python
df['EMA'] = df.ewm(span=20, adjust=False).mean()
df
```

```python
df.plot(title='Close Price')
plt.show()
```
