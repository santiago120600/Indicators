# Simple Moving Average (SMA)

The Simple Moving Average is calculated as:

$$
SMA_t = \frac{P_t+ P_{t-1} + P_{t-2} + \dots + P_{t-n}}{n}
$$

Where:

- $SMA_t$ = Simple Moving Average at time $t$
- $P_t$ = Price (or value) at time $t$
- $n$ = number of total periods
- $P_t + P_{t-1}, \dots$ = previous values

### Example
Last 5 closing prices are: 20, 22, 21, 23, 24

$$
SMA_t =  \frac{20 + 22 + 21 + 23 + 24}{5} =  \frac{110}{5} = 22
$$


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = {'Close': np.random.randint(0, 100, 30).tolist()}
df = pd.DataFrame(data, columns=['Close'])
df
```

```python
df.plot(title='Close Price')
plt.show()
```

```python
df['SMA'] = df.rolling(window=5).mean()
df
```

```python
df.plot(title='Close Price')
plt.show()
```
